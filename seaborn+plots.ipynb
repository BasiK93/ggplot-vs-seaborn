# Boxplot and violin plot

```python
prices=dataset.loc[:,("price", "retail_price")]
```

```python
prices=prices[prices["retail_price"]<100]
```

```python
sns.boxplot(x="variable", y="value", data=pd.melt(prices)).set_title("Ceny produktów w EUR")
plt.show()
```

```python
sns.violinplot(x="variable", y="value", data=pd.melt(prices)).set_title("Ceny produktów w EUR")
plt.show()
```
